```python
def compute_ranks(candidate_summary,S):
  #Sentence tokenization
  generated_summary = sent_tokenize(candidate_summary)
  generated_summary_length = len(generated_summary)
  ranks = list()
  code = 0
  #for each sentence in the summary
  for sentence in generated_summary:
    words = word_tokenize(sentence)
    new_words = list()
    for word in words:
      #Punctuation removal
      if word not in punctuations:
        new_words.append(word)
    sentence = " ".join(new_words)
    #to obtain the position of the sentence in the text
    for key,value in S.items():
      if value == sentence:
        code = key
    rank = 1 - (code/generated_summary_length)
    ranks.append(tuple((sentence, rank)))
  #returns a tuple of sentence and their ranks
  return ranks
```

```python
def reduce_length(generated_summary,maximum_length,ranks):
  sent_list = list()
  for sentence in generated_summary:
    words = word_tokenize(sentence)
    new_words = list()
    for word in words:
      if word not in punctuations:
        new_words.append(word)
    sent_list.append(new_words)
  #print(sent_list)
  model = Word2Vec(sent_list,min_count=1)
  l = list()
  for i in sent_list:
    l.append(vectorizer(i,model))
  X = np.array(l)
  #print(len(X))
  #Creating 2 clusters
  clf = KMeans(n_clusters = 2,max_iter =200,n_init = 1,init="k-means++")
  labels = clf.fit_predict(X)
  print(labels)
  for index,sentence in enumerate(sent_list):
    print(str(labels[index])+":"+str(sentence))
  print("Ranks:",ranks)
  cluster1 = list()
  cluster2 = list()
  count = 0
  for i in labels:
    if i==0:
      cluster1.append(tuple((ranks[count][0],ranks[count][1])))
      count = count+1
    else:
      cluster2.append(tuple((ranks[count][0],ranks[count][1])))
      count = count+1
  #print(cluster1)
  #print(cluster2)
  cluster1 = sorted(cluster1, key=lambda x: x[1],reverse=True)
  #print(cluster1)
  cluster2 = sorted(cluster2, key=lambda x: x[1],reverse=True)
  #print(cluster2)
  count =0
  count1 = 0
  count2 = 0
  final_summary = ""
  #choosing top n/t sentences from each cluster
  while count<maximum_length and count1<len(cluster1) and count2<len(cluster2):
    if cluster1[count1][1]>cluster2[count2][1]:
      final_summary=final_summary+cluster1[count1][0]+". "
      count = count+1
      count1 = count1+1
    else:
      final_summary=final_summary+cluster2[count2][0]+". "
      count = count+1
      count2 = count2+1
  while count<maximum_length and count1<len(cluster1):
    final_summary=final_summary+cluster1[count1][0]+". "
    count = count+1
    count1 = count1+1
  while count<maximum_length and count2<len(cluster2):
    final_summary=final_summary+cluster2[count2][0]+". "
    count = count+1
    count2 = count2+1
  #print(count)
  return final_summary
```

```python
def vectorizer(sent,m):
  vec=[]
  numw = 0
  for w in sent:
    if numw==0:
      vec = m[w]
    else:
      vec = np.add(vec,m[w])
    numw+=1
  return np.asarray(vec)/numw
```

```python
#used for training the TnT tagger using pre-training data
def tag_text(words):
  #obtaining training data from indian module
  train_data = indian.tagged_sents('marathi.pos')
  #print(train_data)
  tnt_pos_tagger = tnt.TnT()
  #Training the tnt Part of speech tagger with marathi data
  tnt_pos_tagger.train(train_data)
  #Tagging input word list using the tnt tagger; output is a list of tuples
  tagged_text = tnt_pos_tagger.tag(words)
  #For each word in the list
  for i in range(len(tagged_text)):
    word = tagged_text[i]
    print(word[0],word[1])
  for i in range(len(tagged_text)):
    word = tagged_text[i]
    #if word is tagged as Unknown
    if word[1]=='Unk':
      #remove the tuple from the list of tuples
      del tagged_text[i]
      #tag the word as a noun
      new_tuple = (word[0],'NN')
      #insert the tuple in the list at the same position to maintain order
      tagged_text.insert(i,new_tuple)
  return tagged_text
```

```python
#Sentence tokenization
sentence_list = sent_tokenize(text)
#Assuming max expected length of summary to be 1/3rd of the original
maximum_length = 31
print(maximum_length)
#Tokenizing title
title_words = word_tokenize(title)
S = dict()
wf = dict()
i =0
#For each sentence in the text
for sentence in sentence_list:
  #Word tokenization
  words = word_tokenize(sentence)
  new_words = list()
  for word in words:
    #Punctuation removal
    if word not in punctuations:
      new_words.append(word)
      #Creating a dictionary containing word-count pairs
      if word in wf.keys():
        wf[word]+=1
      else:
        wf[word] =1
  val = " ".join(new_words)
  #Creating a dictionary containing sentence position-sentence pairs
  S[i] = val
  i = i+1
print(title_words)
print(S)
```

```python
#Initializaing text graph
G = nx.DiGraph()
```

```python
#Algo 1 - Construction of text graph
#Source node for each sentence: S#
source_node = "S#"
edge_label = ""
destination_node = ""
#S = dictionary containing sentence position-sentence pairs
#For each sentence in the text
for key,value in S.items():
  #Word tokenization
  words = word_tokenize(value)
  #POS Tagging words in the sentence - tag_text as defined above
  tags = tag_text(words)
  for tag in tags:
    #if the word is a noun
    if tag[1] == 'NNP' or tag[1] == 'NN':
      word_lemma = tag[0]
      #destination node = current word
      destination_node = word_lemma
      #adding new node representing current word
      if destination_node in wf.keys():
        G.add_node(destination_node,weight=wf[destination_node])
      #Adding edge starting at source node (For first noun in each sentence, source node = S#) and ending at destination node (current word) with weight: Words b/w the nouns
        G.add_edge(source_node,destination_node,weight = edge_label,code = key)
      #Now the current word becomes the source node
        source_node = word_lemma
        edge_label = ""
        destination_node = ""
    #If word is not a noun, add it to the label of the edge connecting the previous noun and next noun
    else:
      edge_label = edge_label+" "+tag[0]
  #Last edge of each sentence = edge connecting last noun in the sentence to E# node
  G.add_edge(source_node,"E#",weight=edge_label,code=key)
  #For beginning of next sentence, source node = S#
  source_node = "S#"
  edge_label = ""
  destination_node = ""
```

```python
for n1,n2,attr in G.edges(data = True):
  print(n1,n2,attr)
```

```python
#Assigning weights to each node in the graph
#wf = dictionary containing word-count pairs
#For each node in the graph
for node in G.nodes:
  if node in wf:
    additional_weight =0
    #if noun is in the title
    if node in title_words:
      count = 0
      for word in title_words:
        if word == node:
          #counting the number of times the noun occurs in the title
          count = count+1
      additional_weight = count*2
    #weight of the node = term frequency in text + (2*term frequency in title)
    G.nodes[node]['weight'] = wf[node]+additional_weight
  #For S# and E#
  else:
    G.nodes[node]['weight'] = 0
#nodes_list = dictionary consisting of nodes and their final weights
nodes_list = dict()
for n,attr in G.nodes(data = True):
  if attr['weight']>0:
    nodes_list[n] = attr
print(nodes_list)
```

```python
source_node_weight =0
dest_node_weight =0
sum =0
med_list =list()
for n,attr in nodes_list.items():
  #sum of weights of all nodes in the graph
  sum+=attr['weight']
  #creating a list of weights of all nodes in the graph
  med_list.append(nodes_list[n]['weight'])
#Finding the average weight of all nodes in the graph
avg_node_weight = sum/len(nodes_list)
#Finding the median weight of all nodes in the graph
median_node_weight = median(med_list)
print(avg_node_weight,median_node_weight)
#source_node_weight = threshold to select source nodes
source_node_weight = max(avg_node_weight,median_node_weight)
print(source_node_weight)
```

```python
#Result graph
C = nx.DiGraph()
```

```python
#Algo 2 - Constructing the result graph
#nodes_list = dictionary consisting of nodes and their final weights
#For each node in the graph
for node,attr in nodes_list.items():
  #if weight of the node is greater than source_weight_threshold, selecct it
  if attr['weight']>=source_node_weight:
    #out_nodes = list of all nodes connected to the source node
    out_nodes = list(G.successors(node))
    print(out_nodes,node)
    #average weight of all neighboring nodes
    average_weight = 0
    #maximum weight of all neighboring nodes
    maximum_weight =0
    #count of all neighbouring nodes excluding S# and E#
    count =0
    #for each neighbouring node
    for out in out_nodes:
      #if node != S# or E#
      if out in nodes_list:
        count = count+1
        average_weight+=nodes_list[out]['weight']
        if nodes_list[out]['weight']>maximum_weight:
          maximum_weight = nodes_list[out]['weight']
    if count!=0:
      average_weight = average_weight/count
    print(average_weight,maximum_weight)
    #destination_node_weight = threshold to select destination nodes: selecting average weight in this case
    destination_node_weight = average_weight
    #for each neighbour in the graph
    for out in out_nodes:
      if out in nodes_list:
        #if weight of the node is greater than or equal to destination_node_weight, select it
        if nodes_list[out]['weight']>=destination_node_weight:
          #adding edge from source node to destination node, code = sentence position
          C.add_edge(node,out,weight =G.get_edge_data(node,out)['weight'],code = G.get_edge_data(node,out)['code'])
print(C.number_of_nodes(),C.number_of_edges())
print(G.number_of_nodes(),G.number_of_edges())
```

```python
#Algo 3 - Selecting sentences
#edge_count_threshold = number of edges needed to be included in result graph for the sentence to be included in summary
edges_count_threshold = 1
candidate_summary = ""
#for each sentence in the text
for i in range(len(S)):
  count = 0
  #iterating over all edges in the graph
  for n1,n2,attr in C.edges(data=True):
    #if edge belongs to sentence i, increment the count
    if attr["code"] == i:
      count = count+1
  #if count>=threshold, include the sentence in the summary
  if count>=edges_count_threshold:
    candidate_summary = candidate_summary+S[i]+". "
print(candidate_summary)
```
